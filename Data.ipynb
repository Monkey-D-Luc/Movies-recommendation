```python
movies['id'] = movies['id'].astype(str)
creadits['id'] = creadits['id'].astype(str)
keywords['id']= keywords['id'].astype(str)
movies = movies.merge(creadits, on='id')
movies =movies.merge(keywords, on ='id')
```

```python
movies = movies[['id','title','genres','keywords','cast','crew']]
movies =movies.iloc[:-25000]
```

```python
movies.dropna (inplace=True)
```

```python
movies = movies.drop_duplicates()
movies = movies.reset_index(drop=True)
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
       if counter < 3:
            L.append(i['name'])
            counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
def convert_title(text):
    L = []
    L = ast.literal_eval(text)
    return L
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['titles'] = movies['title'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['genres']+movies['keywords'] +movies['titles'] +movies['cast']+ movies['crew']
```

```python
new_df = movies[['id','title','tags']]
new_df.head(2)
```

```python
new_df.loc[:,'tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.loc[:,'tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stems(text):
    T = []
    for i in text.split():
        T.append(ps.stem(i))
    return " ".join(T)
```

```python
new_df.loc[:,'tags'] = new_df['tags'].apply(stems)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from scipy.sparse import csr_matrix
cv = CountVectorizer(max_features=3000,stop_words='english')
```

```python
sparse_vector = csr_matrix(cv.fit_transform(new_df['tags']).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
batch_size = 5000
similar_list = []
for i in range(0, len(new_df), batch_size):
    batch_df = new_df.iloc[i:i+batch_size]
    batch_vector = cv.transform(batch_df['tags'])
    sparse_batch_vector = csr_matrix(batch_vector)
    similar_batch = cosine_similarity(sparse_batch_vector, sparse_vector)
    similar_list.append(similar_batch)
similar = np.vstack(similar_list)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similar[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:9]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Superman")
```
